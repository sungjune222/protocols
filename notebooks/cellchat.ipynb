#### CellChat Visualization

```r
H5AD_FILE="macnair_new_new.h5ad"
SPECIES="human"
GROUP_KEY="TCP_oligo"

invisible(
  suppressPackageStartupMessages({
    library(circlize)
    library(dotenv)
    library(rprojroot)
    library(CellChat)
    library(ComplexHeatmap)
  })
)

root <- rprojroot::find_root(rprojroot::has_file("pixi.toml"))
env_path <- file.path(root, ".env")
dotenv::load_dot_env(file = env_ path)

r_dir <- file.path(root, "R")
r_scripts <- list.files(path = r_dir, pattern = "\\.R$", full.names = TRUE, recursive = TRUE)
r_scripts <- r_scripts[!grepl("install\\.R$", r_scripts)]
invisible(sapply(r_scripts, source))

cellchat <- run_cellchat(
  h5ad_file=H5AD_FILE,
  species=SPECIES,
  group_key=GROUP_KEY
)

base_name <- tools::file_path_sans_ext(H5AD_FILE)
cellchat_dir <- get_env_dir("CELLCHAT")
cellchat_dir <- file.path(cellchat_dir, base_name)
```

#### Visualize the specific Ligand-Receptor interaction using a Chord Diagram

```r
# Identified Ligand-Receptor pairs
cellchat@net$LRs
```

```r
interest <- "COL4A2_ITGA3_ITGB1"
cellchat_lr <- subsetCommunication(cellchat, slot.name = "net")
cellchat_lr <- cellchat_lr[order(cellchat_lr$prob, decreasing = TRUE), ]

lr_interest <- cellchat_lr[cellchat_lr$interaction_name == interest, ]
lr_plot <- lr_interest[, c("source", "target", "prob")]

group_cell <- levels(cellchat@idents)
grid_color <- scales::hue_pal()(length(group_cell))
names(grid_color) <- group_cell

min_prob <- min(lr_plot$prob)
max_prob <- max(lr_plot$prob)
col_fun <- colorRamp2(
  breaks = c(min_prob, (min_prob + max_prob)/2, max_prob), 
  colors = c("#0A0F8F", "#FFEE58", "#FF0000") 
)
link_colors <- col_fun(lr_plot$prob)

invisible({
  graphics.off()

  svglite::svglite(
    file.path(cellchat_dir, paste0(interest, "_lr.svg")),
    width = 9, height = 8 
  )
  
  circos.clear()
  circos.par(
    cell.padding = c(0, 0, 0, 0),
    track.margin = c(0.01, 0.01),
    points.overflow.warning = FALSE,
    gap.after = 5
  )

  chordDiagram(
    x = lr_plot,
    grid.col = grid_color,
    col = link_colors,     
    transparency = 0.2,
    directional = 1,
    direction.type = c("diffHeight"), 
    diffHeight = -0.03,
    link.sort = TRUE,
    link.largest.ontop = TRUE,
    annotationTrack = "grid", 
    preAllocateTracks = list(track.height = 0.3)
  )

  circos.track(track.index = 1, panel.fun = function(x, y) {
    circlize::circos.text(
      CELL_META$xcenter, 
      CELL_META$ylim[1], 
      CELL_META$sector.index, 
      
      facing = "clockwise",
      niceFacing = TRUE,
      adj = c(0, 0.5),
      cex = 0.8
    )
  }, bg.border = NA)
  
  lgd <- Legend(
    col_fun = col_fun, 
    title = "Interaction\nStrength", 
    at = c(min_prob, max_prob), 
    labels = c("Low", "High")
  )
  draw(lgd, x = unit(0.9, "npc"), y = unit(0.9, "npc"))

  dev.off()
})
```

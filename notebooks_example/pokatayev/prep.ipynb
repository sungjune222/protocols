```python
import anndata as ad
import gc
import os
import pandas as pd
import scanpy as sc
from scipy.sparse import csr_matrix
from pipeline.utils.env import find_env_dir

raw_data_dir = find_env_dir("RAW_DATA")
pre_h5ad_dir = find_env_dir("PRE_H5AD")

SERIES = "pokatayev"
human_data_dir = os.path.join(raw_data_dir, SERIES, "SCP2927", "anndata")
mouse_data_dir = os.path.join(raw_data_dir, SERIES, "SCP3384", "anndata")
```

#### Human Dataset

```python
HUMAN_FIBROBLAST_FILE = "scp_xenium_human_fib.h5ad"
human_fibroblast_adata = sc.read_h5ad(os.path.join(human_data_dir, HUMAN_FIBROBLAST_FILE))
if not isinstance(human_fibroblast_adata.X, csr_matrix):
    human_fibroblast_adata.X = csr_matrix(human_fibroblast_adata.X)

assert isinstance(human_fibroblast_adata.obs, pd.DataFrame)
human_fibroblast_adata.obs["donor"] = human_fibroblast_adata.obs["donor_id"]
human_fibroblast_adata.obs["sample"] = human_fibroblast_adata.obs["biosample_id"]
human_fibroblast_adata.obs["organ"] = human_fibroblast_adata.obs["organ__ontology_label"]
human_fibroblast_adata.obs["genotype"] = human_fibroblast_adata.obs["genotype"]
human_fibroblast_adata.obs["treatment"] = human_fibroblast_adata.obs["treatment"]
human_fibroblast_adata.obs["celltype"] = human_fibroblast_adata.obs["celltype"]
human_fibroblast_adata.obs["x_axis"] = human_fibroblast_adata.obs["x_centroid"]
human_fibroblast_adata.obs["y_axis"] = human_fibroblast_adata.obs["y_centroid"]
human_fibroblast_adata.obs["slide"] = human_fibroblast_adata.obs["slide_id"]
human_fibroblast_adata.obs["broad_celltype"] = "fibroblast"
human_fibroblast_adata.obs["series"] = SERIES

keep = ["donor", "sample", "organ", "genotype", "treatment", "celltype", "x_axis", "y_axis", "slide", "broad_celltype", "series", ]
human_fibroblast_adata.obs.drop(columns = [c for c in human_fibroblast_adata.obs.columns if c not in keep], inplace=True) #type: ignore

human_fibroblast_adata.obs.head() #type: ignore
```

```python
human_adata_list = [human_fibroblast_adata, human_epithelial_adata, human_stromal_adata, human_lymphoid_adata, human_myeloid_adata, ]
reference_columns = set(human_adata_list[0].obs.columns)

for adata in human_adata_list:
    if set(adata.obs.columns) != reference_columns:
        raise ValueError("obs of adatas do not match!")

human_adata = ad.concat(human_adata_list, join="outer")
del human_adata_list
gc.collect()
```

```python
human_adata.obs.index = human_adata.obs.index.astype(str)
human_adata.var.index = human_adata.var.index.astype(str)
human_adata.write_h5ad(os.path.join(pre_h5ad_dir, SERIES + "_human_raw.h5ad"))
del human_adata
gc.collect()
```
